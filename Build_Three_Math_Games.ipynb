```python
'''Scatter plot game
Randomly generate points on a graph and the player has to input the (x,y) coordinates
For added difficulty, make the graph larger
Algebra practice game
Generate one-step and two-step problems with random integer values and the player has to input the answer
Use positive and negative values. For added difficulty, make the numbers larger
Projectile game
Display a "wall" with random height and location. Player has to move sliders to adjust a parabolic path to clear the wall
For added difficulty, make a second level where players enter a, b, and c without sliders'''
```

```python
#Scatter Plot Game

import matplotlib.pyplot as plt
import random

score = 0

z = int(input("Choose your difficulty level 1-3: "))

if z == 1:
  xmin = -8
  xmax = 8
  ymin = -8
  ymax = 8
elif z == 2:
  xmin = -10
  xmax = 10
  ymin = -10
  ymax = 10
elif z == 3:
  xmin = -15
  xmax = 15
  ymin = -15
  ymax = 15

fig, ax = plt.subplots()

for i in range(0,3):
  xpoint = random.randint(xmin, xmax)
  ypoint = random.randint(ymin, ymax)
  x = [xpoint]
  y = [ypoint]
  plt.axis([xmin,xmax,ymin,ymax])
  plt.plot([xmin,xmax], [0,0], 'b')
  plt.plot([0,0], [xmin,xmax], 'b')
  plt.plot(x,y,'ro')
  print(" ")
  plt.grid()
  plt.show()
  guess = input("Enter the coordinates of the red point: \n")
  guess_array = guess.split(",")
  xguess = float(guess_array[0])
  yguess = float(guess_array[1])

  if guess == xpoint and yguess == ypoint:
    score = score + 1

print("Your score: ", score)
```

```python
#Algebra practice game
import random

def game():
  z1 = int(input("Choose your difficulty level 1-3: "))
  z2 = int(input("1 step or 2 steps? (choose 1 or 2): "))

  if z1 == 1:
    a = random.randint(-100, 100)
    b = random.randint(-100, 100)
    c = random.randint(-100, 100)
  elif z1 == 2:
    a = random.randint(-300, 300)
    b = random.randint(-300, 300)
    c = random.randint(-300, 300)
  elif z1 == 3:
    a = random.randint(-600, 600)
    b = random.randint(-600, 600)
    c = random.randint(-600, 600)

  if z2 == 1:
    question = f"{a} + x = {b}"
    answer = b - a
  elif z2 == 2:
    question = f"{a} + {b} + x = {c}"
    answer = c-(a+b)
  for i in range(1):
    print(question)
    input_ans = int(input("Write your answer: "))
    if input_ans == int(answer):
      print("Your answer is correct.")
    else:
      print("Your answer is incorrect.")

game()
```

```python
#Projectile game
import random
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

wall_dist = random.randint(-10,10)
wall_h = random.randint(1,8)

xmin = -20
xmax = 20
ymin = -20
ymax = 20

points = 2*(xmax-xmin)

linex = [wall_dist, wall_dist]
liney = [0, wall_h]

def f(a,b,c):
  x = np.linspace(xmin, xmax, points)
  y = a*x**2 + b*x + c

  fig, ax = plt.subplots()
  plt.axis([xmin, xmax, ymin, ymax])
  plt.plot([xmin, xmax], [0,0], 'b')
  plt.plot([0,0], [ymin, ymax], 'b')
  plt.plot(linex, liney, 'r')

  plt.plot(x,y)
  plt.show()

interactice_plot = interactive(f, a=(-50,50), b=(-50,50), c=(-50,50))

def manual_values():
  x = np.linspace(xmin, xmax, points)

  fig, ax = plt.subplots()
  plt.axis([xmin, xmax, ymin, ymax])
  plt.plot([xmin, xmax], [0,0], 'b')
  plt.plot([0,0], [ymin, ymax], 'b')
  plt.plot(linex, liney, 'r')

  plt.show()
  a = int(input("Enter value. a= "))
  b = int(input("Enter value. b= "))
  c = int(input("Enter value. c= "))
  y = a*x**2 + b*x + c
  fig, ax = plt.subplots()
  plt.axis([xmin, xmax, ymin, ymax])
  plt.plot([xmin, xmax], [0,0], 'b')
  plt.plot([0,0], [ymin, ymax], 'b')
  plt.plot(linex, liney, 'r')
  plt.plot(x,y)
  plt.show


z = int(input("choose difficulty level 1 or 2: "))
if z == 1:
  interactice_plot
elif z==2:
  manual_values()
```
